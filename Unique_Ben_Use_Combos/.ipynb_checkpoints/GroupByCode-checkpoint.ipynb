```python
# SQL Results of AllocationAmounts Unique Benefical Use Combinations
fileInput = "Allocation_Benuse.csv"
dfload = pd.read_csv(fileInput)
print(len(dfload))
dfload.head(1)
```

```python
# Joining df on duplicates based on key fields.
df = dfload.copy()
df = df.replace(np.nan, "")  # Replaces NaN values with blank.
groupbyList = ['State', 'AllocationNativeID']
df = df.groupby(groupbyList).agg(lambda x: ','.join([str(elem) for elem in (list(set(x))) if elem!=''])).replace(np.nan, "").reset_index()
# df = df[columnslist]  # reorder the dataframe's columns based on columnslist
print(len(df))
df.head(1)
```

```python
# Counting number of CFS, AF, and WaDEBenUse
def CountNumItems(val):
    val = str(val).strip()
    valList = val.split(",")
    outString = len(valList)
    return outString

df['Count_CFS'] = df.apply(lambda row: CountNumItems(row['AllocationFlow_CFS']), axis=1)
df['Count_AF'] = df.apply(lambda row: CountNumItems(row['AllocationVolume_AF']), axis=1)
df['Count_WBU'] = df.apply(lambda row: CountNumItems(row['WaDEBenUse']), axis=1)
df.head()
```

```python
# sort CFS, AF, and WaDEBenUse
def CountNumItems(val):
    val = str(val).strip()
    valList = val.split(",")
    valList = sorted(valList)
    outString = ",".join(str(x) for x in valList)
    return outString

# df['AllocationFlow_CFS'] = df.apply(lambda row: CountNumItems(row['AllocationFlow_CFS']), axis=1)
# df['AllocationVolume_AF'] = df.apply(lambda row: CountNumItems(row['AllocationVolume_AF']), axis=1)
df['WaDEBenUse'] = df.apply(lambda row: CountNumItems(row['WaDEBenUse']), axis=1)
df.head()
```
